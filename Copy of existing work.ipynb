```python
sorted_df['Date'] = pd.to_datetime(sorted_df['transaction_date']).dt.date

pivot_df = sorted_df.pivot_table(index='Date', columns='sub_sku', values='sell_qty', aggfunc='sum')

pivot_df.reset_index(inplace=True)

pivot_df.columns
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import matplotlib.pyplot as plt

# Assuming main_df contains the relevant columns for prediction
# 'transaction_date' is the date column, and multiple 'item' columns are the target columns you want to predict

# Example: Select relevant columns
df = pivot_df.copy()
df['Date'] = pd.to_datetime(df['Date'])

# Extract features (if any) from the date, e.g., day of week, month, etc.
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year

# Create a new DataFrame to store future predictions
future_predictions_df_forest = pd.DataFrame({'Date': pd.date_range(df['Date'].max(), periods=60, freq='D')})

# Loop over each item column
for item_column in df.columns[1:441]:  # Add more columns as needed
    print(f"Processing {item_column}...")

    # Check for and handle missing values in the target variable 'y'
    df[item_column].fillna(0, inplace=True)

    # Select features and target
    features = ['day_of_week', 'month', 'year']  # Add more features if needed
    target = item_column

    # Use all data for training (no separate test set)
    X_train, y_train = df[features], df[target]

    # Initialize and train a Random Forest regressor
    model = RandomForestRegressor(n_estimators=11,random_state=42, max_depth=2)
    model.fit(X_train, y_train)

    # Make predictions for future dates
    future_dates = pd.date_range(df['Date'].max(), periods=60, freq='D')
    future_features = pd.DataFrame({'day_of_week': future_dates.dayofweek, 'month': future_dates.month, 'year': future_dates.year})
    future_predictions = model.predict(future_features)

    # Store future predictions in the DataFrame
    future_predictions_df_forest[item_column] = future_predictions

# Export the DataFrame to an Excel sheet
#excel_file_path = 'future_predictions_forest_regression.xlsx'
#future_predictions_df.to_excel(excel_file_path, index=False)
#print(f"Excel sheet saved to {excel_file_path}")
```

```python
k=0
diff_tot=0
forest=[]
for i in future_predictions_df_forest.columns[1:-1]:
  forecast = future_predictions_df_forest[i][8:13].sum().round()
  actual = df_actual['Actual Sold'][k]
  forest=np.append(forest,forecast - actual)
  k += 1
  diff_tot += (forecast - actual)**2
  print(i,forecast - actual)
print('MSE',diff_tot/len(future_predictions_df_forest.columns[1:-1]))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

# Assuming main_df contains the relevant columns for prediction
# 'transaction_date' is the date column, and multiple 'item' columns are the target columns you want to predict

# Example: Select relevant columns
df = pivot_df.copy()
df['Date'] = pd.to_datetime(df['Date'])

# Create a new DataFrame to store future predictions
future_predictions_df_linear = pd.DataFrame({'Date': pd.date_range(df['Date'].max(), periods=60, freq='D')})

# Loop over each item column
for item_column in df.columns[1:441]:  # Add more columns as needed
    print(f"Processing {item_column}...")

    # Check for and handle missing values in the target variable 'y'
    df[item_column].fillna(0, inplace=True)

    # Select the time series for the current item
    time_series = df[item_column].values.reshape(-1, 1)

    # Scale the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(time_series)

    # Create sequences for training the linear regression model
    def create_sequences(data, seq_length):
        sequences = []
        for i in range(len(data) - seq_length):
            seq = data[i:i+seq_length]
            label = data[i+seq_length:i+seq_length+1]
            sequences.append((seq, label))
        return np.array(sequences)

    seq_length = 7  # You can adjust this based on your preference
    sequences = create_sequences(scaled_data, seq_length)

    # Split the data into features and labels
    X = np.array([seq[0].flatten() for seq in sequences])
    y = np.array([seq[1].flatten()[0] for seq in sequences])

    # Initialize and train a Linear Regression model
    model = LinearRegression()
    model.fit(X, y)

    # Make predictions for future dates
    future_dates = pd.date_range(df['Date'].max(), periods=60, freq='D')
    future_features = np.array([scaled_data[-seq_length:].flatten() for _ in range(60)])
    future_predictions_scaled = model.predict(future_features)

    # Inverse transform to get back the original scale
    future_predictions = scaler.inverse_transform(future_predictions_scaled.reshape(-1, 1))

    # Store future predictions in the DataFrame
    future_predictions_df_linear[item_column] = future_predictions

# Export the DataFrame to an Excel sheet
#excel_file_path = 'future_predictions_linear_regression.xlsx'
#future_predictions_df.to_excel(excel_file_path, index=False)
#print(f"Excel sheet saved to {excel_file_path}")
```

```python
def my_func(s):
  if s==[]:
    return s
  else:
    v=[]
    k=0
    for i in s:
      if i =="C":
        v=v[0:-1]
      elif i =="+":
        v.append(sum(v[k-2:]))
      elif i =="D":
        v.append(2*v[k-1])
      else:
        v.append(int(i))
  return sum(v)
```
