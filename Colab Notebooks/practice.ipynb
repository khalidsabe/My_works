```python
k=[2]*7
```

```python
k.sort()
l=len(k)
p=-1
for i in range(l,0,-1):
  if k[l-i] >= i:
    p=i
    break;
if p == -1:
  print('not found')
else:
  print(f'there are', p, 'integers greater than or equal to', p)
```

```python
s='abnsyuvdbASDKmvdcgvcjhbkvkuyvgd,jfhnnn'
```

```python
vocab={}
for i in s:
  if i.isalpha():
    if i in vocab.keys():
      vocab[i]+=1
    else:
      vocab[i]=1
for i,v in vocab.items():
  if v==max(vocab.values()):
    print(i)
```
