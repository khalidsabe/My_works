```python
import pandas as pd
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.preprocessing import MinMaxScaler

# Assuming main_df contains the relevant columns for prediction
# 'transaction_date' is the date column, and multiple 'item' columns are the target columns you want to predict

# Example: Select relevant columns
df = pivot_df.copy()
df['Date'] = pd.to_datetime(df['Date'])

# Create a new DataFrame to store future predictions
future_predictions_df_g = pd.DataFrame({'Date': pd.date_range(df['Date'].max(), periods=60, freq='D')})

# Loop over each item column
for item_column in df.columns[1:441]:  # Add more columns as needed
    print(f"Processing {item_column}...")

    # Check for and handle missing values in the target variable 'y'
    df[item_column].fillna(0, inplace=True)

    # Select the time series for the current item
    time_series = df[item_column].values.reshape(-1, 1)

    # Scale the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(time_series)

    # Create sequences for training the Gaussian Process model
    def create_sequences(data, seq_length):
        sequences = []
        for i in range(len(data) - seq_length):
            seq = data[i:i+seq_length]
            label = data[i+seq_length:i+seq_length+1]
            sequences.append((seq.flatten(), label.flatten()))
        return np.array(sequences)

    seq_length = 6  # You can adjust this based on your preference
    sequences = create_sequences(scaled_data, seq_length)

    # Split the data into features and labels
    X = np.array([seq[0] for seq in sequences])
    y = np.array([seq[1] for seq in sequences])

    # Define the Gaussian Process kernel
    kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))

    # Initialize and train a Gaussian Process regressor
    model = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
    model.fit(X, y)

    # Make predictions for future dates
    future_dates = pd.date_range(df['Date'].max(), periods=60, freq='D')
    future_features = np.array([scaled_data[-seq_length:].flatten() for _ in range(60)])
    future_predictions_scaled, _ = model.predict(future_features, return_std=True)

    # Inverse transform to get back the original scale
    future_predictions = scaler.inverse_transform(future_predictions_scaled.reshape(-1, 1))

    # Store future predictions in the DataFrame
    future_predictions_df_g[item_column] = future_predictions
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import GRU, Dense
from keras.optimizers import Adam

# Assuming main_df contains the relevant columns for prediction
# 'transaction_date' is the date column, and multiple 'item' columns are the target columns you want to predict

# Example: Select relevant columns
df = pivot_df.copy()
df['Date'] = pd.to_datetime(df['Date'])

# Extract features (if any) from the date, e.g., day of week, month, etc.
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year

# Create a new DataFrame to store future predictions
future_predictions_df = pd.DataFrame({'Date': pd.date_range(df['Date'].max(), periods=60, freq='D')})

# Loop over each item column
for item_column in df.columns[1:441]:  # Add more columns as needed
    print(f"Processing {item_column}...")

    # Check for and handle missing values in the target variable 'y'
    df[item_column].fillna(0, inplace=True)

    # Select the time series for the current item
    time_series = df[item_column].values.reshape(-1, 1)

    # Scale the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(time_series)

    # Create sequences for training the GRU model
    def create_sequences(data, seq_length):
        sequences = []
        for i in range(len(data) - seq_length):
            seq = data[i:i+seq_length]
            label = data[i+seq_length:i+seq_length+1]
            sequences.append((seq, label))
        return np.array(sequences)

    seq_length = 6  # You can adjust this based on your preference
    sequences = create_sequences(scaled_data, seq_length)

    # Split the data into features and labels
    X = np.array([seq[0] for seq in sequences])
    y = np.array([seq[1] for seq in sequences])

    # Build the GRU model
    model = Sequential()
    model.add(GRU(50, activation='relu', input_shape=(seq_length, 1)))
    model.add(Dense(1))
    model.compile(optimizer=Adam(), loss='mean_squared_error')

    # Train the model
    model.fit(X, y, epochs=50, batch_size=32, verbose=0)

    # Make predictions for future dates
    last_sequence = scaled_data[-seq_length:]
    future_predictions_scaled = []

    for _ in range(60):
        pred = model.predict(last_sequence.reshape(1, seq_length, 1))
        future_predictions_scaled.append(pred[0, 0])
        last_sequence = np.append(last_sequence[1:], pred[0, 0])

    # Inverse transform to get back the original scale
    future_predictions = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1, 1))

    # Store future predictions in the DataFrame
    future_predictions_df[item_column] = future_predictions

# Export the DataFrame to an Excel sheet
#excel_file_path = 'future_predictions_gru.xlsx'
#future_predictions_df.to_excel(excel_file_path, index=False)
#print(f"Excel sheet saved to {excel_file_path}")
```
